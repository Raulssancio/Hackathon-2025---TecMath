```python
import pandas as pd

# Load your CSV (replace path with your actual file)
df = pd.read_csv('Temperatura del aire cerca del suelo.csv')

# Convert 'FechaHora' to datetime
df['FechaHora'] = pd.to_datetime(df['FechaHora'])

# Extract components
df['Year'] = df['FechaHora'].dt.year
df['Month'] = df['FechaHora'].dt.month
df['Day'] = df['FechaHora'].dt.day
df['Hour'] = df['FechaHora'].dt.hour

# Resulting DataFrame contains separated columns
print(df.head())

```

```python
# Assuming previous steps have been done and df has 'year', 'month', 'day' columns
filtered_month_8 = df[(df['Month'] == 8) & (df['Day'] == 8) & (df['Hour'] == 8)]

# Display only the temperatures from this selection
print(filtered_month_8[['FechaHora', 'Temp']])
filtered_month_8.head(10)

```

```python
filtered_month_8.plot(x='Year', y='Temp', kind='line', title='Temperature on August 8 (All Years)')
```

# Available Plot Types in Pandas

Common `kind` options for pandas plotting:
- 'line': Line plot (default) - good for time series
- 'scatter': Scatter plot - good for seeing individual points
- 'bar': Bar plot - good for categorical data
- 'hist': Histogram - good for distribution of values
- 'box': Box plot - good for statistical summaries
- 'area': Area plot - good for showing cumulative effect

```python
# Example of different plot types
import matplotlib.pyplot as plt

# Create a figure with multiple subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# Line plot
filtered_month_8.plot(x='FechaHora', y='Temp', kind='line', ax=ax1, title='Line Plot')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)

# Scatter plot
filtered_month_8.plot(x='FechaHora', y='Temp', kind='scatter', ax=ax2, title='Scatter Plot')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)

# Box plot
filtered_month_8.boxplot(column='Temp', ax=ax3, title='Box Plot')

# Histogram
filtered_month_8['Temp'].plot(kind='hist', bins=20, ax=ax4, title='Histogram')

plt.tight_layout()
plt.show()
```
